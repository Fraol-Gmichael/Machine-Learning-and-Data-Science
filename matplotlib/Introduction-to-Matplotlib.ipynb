## Matplotlib example Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare the data 
x = [1, 2, 3, 4]
y = [11, 2, 3, 44]

# 2. setup the plot
fig, ax = plt.subplots(figsize=(10, 10)) #width, height

# 3. plot the data
ax.plot(x, y)

# 4. customise a plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show (Save the whole figure)
fig.savefig("./images/sample-plot.png")
```

## Making figures with numpy arrays

We can do:
* Line plot
* Scatter plot
* Bar plot

```python
import numpy as np
```

```python
# create some data
x = np.linspace(0, 10, 100)
```

```python
# plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# use same data to make a scatter 
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
nut_butter_prices = {"Almond Butter": 10,
                    "Peanut Butter": 8,
                    "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store", 
      ylabel="Price ($)");
```

## Histograms and Subplots

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
ax.set(title="Nut Butter Store", 
      xlabel="Price ($)");
```

```python
# make a normal distribution
num = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(num);
```

```python
# subplots option 1
fig, ((ax1, ax2, ax3),(ax4, ax5, ax6)) = plt.subplots(nrows=2,
                                             ncols=3,
                                             figsize=(10, 5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
fig, ax = plt.subplots(nrows=2, 
                       ncols=3, 
                       figsize=(10, 5))

# plot to each different axis
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[0, 2].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 0].hist(np.random.randn(1000));
```

```python
heart_disease = pd.read_csv('heart-disease.csv')
heart_disease.describe()
```

```python
heart_disease.info()
```

```python
heart_disease['age'].plot.hist(bins=10);
```

```python
heart_disease.plot.hist(figsize=(10, 30), subplots=True);
```

```python
age_over_50 = heart_disease[heart_disease['age'] > 50]
age_over_50.head()
```

```python
age_over_50.plot(kind="scatter",
                 x='age',
                 y='chol',
                 c='target');
```

```python
## oo method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
age_over_50.plot(kind = 'scatter',
                x = 'age',
                y = 'chol',
                c = 'target', 
                ax = ax);

# ax.set_xlim([45, 100])
```

```python
## Pure oo method
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=age_over_50['age'],
                    y=age_over_50['chol'],
                    c=age_over_50['target']);

ax.set(title='Heart Disease and Cholestrol Levels',
      xlabel="Age", 
      ylabel="Cholestrol");
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
# add a horizontal line
ax.axhline(age_over_50["chol"].mean(), linestyle='--', color='red');
```

### create subplots of pandas dataframe

```python
plt.style.use("seaborn-whitegrid")
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                              sharex=True);

scatter = ax0.scatter(x=age_over_50['age'],
                    y=age_over_50['chol'],
                    c=age_over_50['target'],
                    cmap='winter');
# add a legend
ax0.legend(*scatter.legend_elements(), title="Target");

# customize ax0
ax0.set(title='Heart Disease and Cholestrol Levels',
       ylabel='Cholestrol')

# add a horizontal line
ax0.axhline(age_over_50["chol"].mean(), linestyle='--', color='red');

scatter = ax1.scatter(x=age_over_50['age'],
                    y=age_over_50['thalach'],
                    c=age_over_50['target'],
                    cmap='winter')
# add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# add a horizontal line
ax1.axhline(age_over_50["thalach"].mean(), linestyle='--', color='red');

# customize ax1
ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate')

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold');

fig.savefig("exported image.png")
```
